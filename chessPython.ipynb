```
def states_after_legal_moves(board):
  legal_moves = [str(move) for move in list(board.legal_moves)]
  future_boards = []
  for move in legal_moves:
    move = chess.Move.from_uci(move)
    board.push(move)

    temp_board = [list(row) for row in str(board).replace(" ", "").split("\n")]
    future_boards.append(temp_board)

    board.pop()

  return future_boards
```

```
future = states_after_legal_moves(board)
```

```
def remove_all_not_pawn_or_king(board):
  arr = ["p", "P", "k", "K"]
  temp_board = [list(row) for row in str(board).replace(" ", "").split("\n")]
  for i in range(len(temp_board)):
    for j in range(len(temp_board[i])):
      if temp_board[i][j] not in arr:
        temp_board[i][j] = "."

  return temp_board
```

```
remove_all_not_pawn_or_king(board)
```
